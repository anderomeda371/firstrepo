```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("/content/train.csv")
dataset.head
dataset=dataset.dropna(axis=0)
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
y=y.reshape((-1,1))
X=X.reshape((-1,1))
print(X.shape,y.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
X_standard=StandardScaler()
y_standard=StandardScaler()

X_s=X_standard.fit_transform(X.reshape(-1,1))
y_s=y_standard.fit_transform(y)

regressor=LinearRegression()

regressor.fit(X_s,y_s)

min(X_s),min(y_s)
```

```python
plt.scatter(X_s,y_s,color="blue",s=5)
```

```python
test_set=pd.read_csv("/content/test.csv")
test_set=test_set.dropna(axis=0)
test_set.info()

Xt=test_set.iloc[:,:-1].values
yt=test_set.iloc[:,-1].values

Xt_s=X_standard.transform(Xt.reshape(-1,1))
yt_s=y_standard.transform(yt.reshape(-1,1))
max(yt_s)

```

```python
Y_predict=regressor.predict(Xt_s)
plt.plot(Xt_s,Y_predict,c="green")
plt.scatter(X_s,y_s,color="blue",s=5)
plt.scatter(Xt_s,yt_s,color="red",s=5)
plt.show()
```

```python
from sklearn.metrics import d2_absolute_error_score

accuracy=d2_absolute_error_score(Y_predict,yt)
accuracy
```
