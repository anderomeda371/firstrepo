```python
def plot_count(x,fig):
  # plt.subplot(4,2,fig)
  # plt.title(fig)
  # sns.countplot(data=dataset,x=dataset[X])
  # # plt1=dataset[X].value_counts().plot(kind="bar")
  # plt.subplot(4,2,fig+1)
  # plt.title(fig)
  # sns.boxplot(x=dataset[X],y=dataset.price)
  # df=pd.DataFrame(dataset.groupby(X)["price"].mean())
  # # df.plot(kind="bar")
    plt.subplot(4,2,fig)
    plt.title(x+' Histogram')
    # sns.countplot(dataset[x],palette=("magma"))
    dataset[x].value_counts().plot(kind="bar")
    plt.subplot(4,2,(fig+1))
    plt.title(x+' vs Price')
    sns.boxplot(x=dataset[x], y=dataset.price, palette=("magma"))

plt.figure(figsize=(15,20))
plot_count("enginelocation",3)
plot_count("cylindernumber",1)
plot_count("fuelsystem",5)
plot_count("drivewheel",7)
```

```python
from sqlalchemy import label
temp=dataset.copy()
# dataset["price"]=dataset["price"].astype(int)
table=dataset.groupby("CompanyName")["price"].mean()
temp=temp.merge(table.reset_index() ,how="left" ,on="CompanyName")
temp
bins=[0,10000,20000,40000]
car_bins=["Budget","Medium","High"]
dataset["carsRange"]=pd.cut(temp["price_y"],bins,right=False,labels=car_bins)
dataset.head()
```

```python
final_set=dataset[['price', 'fueltype', 'aspiration','carbody', 'drivewheel','wheelbase',
                  'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower',
                    'fueleconomy', 'carlength','carwidth', 'carsRange']]
```

```python
def dummyMaker(x,dset):
  temp=pd.get_dummies(dset[x],drop_first=True)
  df=pd.concat([dset,temp],axis=1)
  df.drop([x],axis=1,inplace=True)
  return df

final_set=dummyMaker("fueltype",final_set)
dummyMaker("cylindernumber",final_set)
dummyMaker("carbody",final_set)
dummyMaker("aspiration",final_set)
dummyMaker("drivewheel",final_set)

final_set


```
